```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
training_data=pd.read_csv(r"C:\Users\ojoja\Desktop\works\DSML project 1\train_data.csv", header=None)
training_data.head(5)
```

```python
training_data = training_data.drop(labels=0, axis=0)
training_data
```

```python
training_data.isnull().values.any()
```

```python
training_data = training_data.astype(float)
training_data.head(5)
```

```python
training_data_classes=pd.read_csv(r"C:\Users\ojoja\Desktop\works\DSML project 1\train_labels.csv", header=None)
training_data_classes.head(5)
```

```python
training_data_classes = training_data_classes.drop(labels=0, axis=0)
training_data_classes.head(5)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
x = training_data.iloc[:,0:]
y = training_data_classes.iloc[:,1:].values.ravel()
# split the dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report

nb = GaussianNB().fit(np.array(X_train), y_train)
pred_train = nb.predict(np.array(X_test))
print(classification_report(y_test, pred_train))
```

```python
nb_tune = GridSearchCV(nb, param_grid={'var_smoothing': [0.00000001, 0.000000001, 0.00000001]},cv=5, scoring='accuracy').fit(np.array(x), y)
pred_train = nb_tune.predict(np.array(x))
print(classification_report(y, pred_train))
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression().fit(np.array(X_train), y_train)
pred_train = lr.predict(np.array(X_test))
print(classification_report(y_test, pred_train))
```

```python
# # importing cross-validation from sklearn package.
# from sklearn.model_selection import cross_val_score,KFold
  
# kf=KFold(n_splits=5)
# score=cross_val_score(lr,x,y,cv=kf)
# print("Cross Validation Scores are {}".format(score))
# print("Average Cross Validation score :{}".format(score.mean()))
```

```python
lr_tune = GridSearchCV(lr, param_grid={'C': [0.01,0.1,1]},cv=5, scoring='accuracy').fit(np.array(x), y)
pred_train = lr_tune.predict(np.array(x))
print(classification_report(y, pred_train))
```

## K-Nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier().fit(np.array(X_train), y_train)
pred_train = knn.predict(np.array(X_test))
print(classification_report(y_test, pred_train))
```

```python
# # importing cross-validation from sklearn package.
# from sklearn.model_selection import cross_val_score,KFold
  
# kf=KFold(n_splits=5)
# score=cross_val_score(knn,x,y,cv=kf)
# print("Cross Validation Scores are {}".format(score))
# print("Average Cross Validation score :{}".format(score.mean()))
```

```python
knn_tune = GridSearchCV(knn, param_grid={'n_neighbors': [5,10,15], 'leaf_size': [5, 10],},cv=5, scoring='accuracy').fit(np.array(x), y)
pred_train = knn_tune.predict(np.array(x))
print(classification_report(y, pred_train))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 100) 
clf.fit(np.array(X_train), y_train) 
pred_train = clf.predict(np.array(X_test))
print(classification_report(y_test, pred_train))
```

```python
# importing cross-validation from sklearn package.
from sklearn.model_selection import cross_val_score,KFold
  
kf=KFold(n_splits=5)
score=cross_val_score(clf,x,y,cv=kf)
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

```python
clf_tune = GridSearchCV(clf, param_grid={'n_estimators': [5,100,500], 'max_leaf_nodes': [ 10, 'None']},cv=5, scoring='accuracy').fit(np.array(x), y)
pred_train = clf_tune.predict(np.array(x))
print(classification_report(y, pred_train))
```
